```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import plotly.express as px

# Outer Circle
theta = np.linspace(0, 2*np.pi, 200)
r_outer = 5

x_outer = r_outer * np.cos(theta)
y_outer = r_outer * np.sin(theta)

# Inner Circle
r_inner = 3

x_inner = r_inner * np.cos(theta)
y_inner = r_inner * np.sin(theta)

# Plot circles
plt.scatter(x_outer, y_outer, color="blue")
plt.scatter(x_inner, y_inner, color="red")
plt.axis("equal")
plt.show()

# Create dataset
X = np.vstack([
    np.column_stack([x_outer, y_outer]),
    np.column_stack([x_inner, y_inner])
])

# Labels
Y = np.hstack([
    np.zeros(len(x_outer)),  # outer circle
    np.ones(len(x_inner))    # inner circle
])

# DataFrame
df = pd.DataFrame(X, columns=["X1", "X2"])
df["Y"] = Y

# Kernel transformation
df["X1*X2"] = df["X1"] * df["X2"]

# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    df[["X1", "X2"]], df["Y"], test_size=0.25, random_state=42
)

# 3D Visualization
```

```python
fig = px.scatter_3d(
    df,
    x="X1",
    y="X2",
    z="X1*X2",
    color="Y"
)

fig.show()
```
